```python
import math
import csv

x = [1, 0]
y = [0, 1]
learning_rate = 0.1
w = 0.5000
g = 0.5000

def sigmoid(z):
  return 1 / (1 + math.exp(-z))
def sigmoid_derivative(yhat):
  return yhat * (1 - yhat)
with open("625-01.csv", "w", newline='') as file:
  writer = csv.writer(file)
  writer.writerow(["Epoch", "x", "y", "w", "g", "Output"])

  for i in range(925):
    for j in range(2):
      z = w * x[j] + g
      yhat = sigmoid(z)
      E = 0.5 * ((y[j] - yhat) ** 2)
      dE = yhat - y[j]
      dE_dW = dE * sigmoid_derivative(yhat) * x[j]
      dE_dG = dE * sigmoid_derivative(yhat)
      w -= learning_rate * dE_dW
      g -= learning_rate * dE_dG
      output = 1 if w * x[j] + g > 0 else 0
      print(f"Iteration {i+1} (x={x[j]}): w = {w:.4f}, g = {g:.4f}, actual value = {w * x[j] + g:.4f}, output = {output}")
      writer.writerow([i+1, x[j], y[j], w, g, output])
```
